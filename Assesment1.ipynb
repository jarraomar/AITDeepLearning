```python
import cv2
# Load the image
img = cv2.imread('straw-hat-flag.jpg')

# Get the shape of the image
h, w, c = img.shape

# Define the region to change
x, y = w - 30, 0
w, h = 30, 30

# Create a white rectangle
img[y:y+h, x:x+w] = (0, 0, 0)

# Save the modified image
cv2.imwrite('modified_image.jpg', img)
```
